```python
# Ready for the regression!
```

```python
y = dataFrame["price"].values
x = dataFrame.drop("price",axis=1).values
# Convert to numpy series.
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
# 9090 datas adnd 5 features. Removed "transmission" and "price" added to y. 
```

```python
model = Sequential()

model.add(Dense(12,activation="relu")) # 12 neurons hidden layer
model.add(Dense(12,activation="relu")) # 12 neurons hidden layer
model.add(Dense(12,activation="relu")) # 12 neurons hidden layer
model.add(Dense(12,activation="relu")) # 12 neurons hidden layer

model.add(Dense(1)) # 1 neuron output layer
```

```python
model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x = x_train, y = y_train, validation_data = (x_test, y_test), batch_size = 250, epochs=300)
# validate while train (validation_data)
# piece by piece training 
```

```python
lostData = pd.DataFrame(model.history.history)
```

```python
lostData.head()
```

```python
lostData.plot() # loss and val_loss have to end together in the graphic.
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizisi = model.predict(x_test)
```

```python
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi)
```
